```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
experience = np.array([1,2,3,4,5,6]).reshape(-1,1)
salary = np.array([20000,30000,40000,50000,60000,50000])

# Model
model = LinearRegression()
model.fit(experience, salary)

# Prediction
print("Salary for 7 years:", model.predict([[7]]))
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier

# Data
age = [[18],[22],[25],[30],[35]]
buy = [0,0,1,1,1]

# Model
model = DecisionTreeClassifier()
model.fit(age, buy)

# Prediction
print("Prediction for age 28:", model.predict([[28]]))
```
